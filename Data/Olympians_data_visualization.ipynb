# Data Cleaning

```python
df_india.drop(['ID', 'NOC','Team', 'Games', 'Event'], axis=1, inplace=True)
df_india.head()
```

```python
df_india.isnull().sum()
```

```python
df_india['Age'].fillna(value=df_india.Age.mean(), inplace=True)
df_india['Height'].fillna(value=df_india.Height.mean(), inplace=True)
df_india['Weight'].fillna(value=df_india.Weight.mean(), inplace=True)
df_india.isnull().sum()
```
